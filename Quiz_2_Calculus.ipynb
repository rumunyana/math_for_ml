````python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
````

````python
def derivative_function(func_str):
    """
    Calculate the  derivative of a given function with respect to variable x.

    Parameters:
    func_str (str): The function you want to find the derivative of, written in text.
                    For example, if you want the derivative of x^2, you write "x**2".

    Returns:
    The derivative (calculated result) of the function you provided.
    """

    # Define the symbolic variable
    x = sp.symbols('x')

    # Parse the input string to a SymPy expression
    func = sp.sympify(func_str)

    # Calculate the derivative
    derivative = sp.diff(func, x)

    return derivative
````

# Task 4: Create separate array of Minimas and maximas

````python
import sympy as sp

def find_critical_points_and_coordinates(f_str, variable_str='x'):
    x = sp.symbols(variable_str)
    f = sp.sympify(f_str)

    # Calculate first and second derivatives
    f_prime = sp.diff(f, x)
    f_double_prime = sp.diff(f_prime, x)

    # Find critical points where the first derivative is zero
    critical_points = sp.solve(f_prime, x)

    # Calculate (x, y) coordinates for each critical point
    critical_points_coordinates = [(point, f.subs(x, point)) for point in critical_points]

    # Classify each critical point
    minima = [(x_val, y_val) for x_val, y_val in critical_points_coordinates if f_double_prime.subs(x, x_val) > 0]
    maxima = [(x_val, y_val) for x_val, y_val in critical_points_coordinates if f_double_prime.subs(x, x_val) < 0]

    return minima, maxima

# Example function
f_str = "x**2 - 2*x + 1"

# Find and classify critical points with coordinates
minima_coordinates, maxima_coordinates = find_critical_points_and_coordinates(f_str)
print('Minima (x, y):', minima_coordinates)
print('Maxima (x, y):', maxima_coordinates)
````

````python
def find_global_extrema_with_coordinates(f_str, variable_str='x'):
    # Use the function from Task 4
    minima, maxima = find_critical_points_and_coordinates(f_str, variable_str)

    # Find global minimum and maximum
    global_min = min(minima, key=lambda item: item[1], default=(None, None))
    global_max = max(maxima, key=lambda item: item[1], default=(None, None))

    return global_min, global_max

# Find global minima and maxima with coordinates
global_min_coordinates, global_max_coordinates = find_global_extrema_with_coordinates(f_str)
print('Global Minimum (x, y):', global_min_coordinates)
print('Global Maximum (x, y):', global_max_coordinates)
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Function to plot the quadratic equation and its global minima
def plot_quadratic_with_extrema(f_str, global_min, variable_str='x', range_min=-10, range_max=10):
    x = sp.symbols(variable_str)
    f = sp.sympify(f_str)

    # Lambdify the function for numerical evaluation
    f_lambdified = sp.lambdify(x, f, 'numpy')

    # Generate x values and calculate y values
    x_vals = np.linspace(range_min, range_max, 400)
    y_vals = f_lambdified(x_vals)

    # Plot the quadratic function
    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, label=f_str)

    # Plot the global minimum if it exists
    if global_min[0] is not None:
        plt.scatter(*global_min, color='red', zorder=5)
        plt.annotate(f'Global Min ({global_min[0]}, {global_min[1]})', xy=global_min, xytext=(global_min[0]+1, global_min[1]),
                     arrowprops=dict(facecolor='black', arrowstyle='->'), zorder=5)

    plt.title(f'Plot of finding global minima and maxima')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Redefine the function "x**2 - 2*x + 1"
new_f_str = "x**2 - 2*x + 1"

# Find global minima
new_global_min_coordinates, _ = find_global_extrema_with_coordinates(new_f_str)

# Plot the function with its global minima
plot_quadratic_with_extrema(new_f_str, new_global_min_coordinates, range_min=-2, range_max=4)

````
